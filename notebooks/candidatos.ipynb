# Candidatos das Eleições Municipais - São Paulo

```python
import os
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import IFrame

%matplotlib notebook
```

```python
from eleicoes.candidatos import *
```

```python
UF = 'SP'
MUNICIPIO = 71072
```

```python
ANOS = (2012, 2016)
pleitos = {}
for ano in ANOS:
    pleitos[ano] = dados_candidatos(ano, UF, MUNICIPIO)
```

```python
pleitos[2012].columns
```

Seleciona apenas os cargos de prefeito e vice prefeito

```python
PREFEITO = 11
VICE_PREFEITO = 12
```

```python
colunas = ['ANO_ELEICAO', 'NUM_TURNO', 'COD_LEGENDA', 'SIGLA_PARTIDO', 'NOME_CANDIDATO', 'DESC_CARGO',
           'COD_SIT_TOT_TURNO', 'DESPESA_MAX_CAMPANHA','DESC_SIT_TOT_TURNO']
for ano in pleitos:
    pleitos[ano] = (pleitos[ano].query('COD_CARGO in (@PREFEITO, @VICE_PREFEITO)')
                    .sort_values(['NUM_TURNO', 'COD_LEGENDA','DESC_CARGO'])
                    .reset_index(drop=True))
```

```python
pleitos[2012]
```

```python
pd.options.display.float_format = '{:,.2f}'.format
pleitos[2016]
```

# Cria dataset único para análise dos pleitos

```python
eleicoes = pd.concat(pleitos[df] for df in pleitos)
```

```python
primeiro_turno = eleicoes.query('NUM_TURNO == 1')
```

```python
TITULO = 'Qtd. de Partidos'
primeiro_turno.groupby('ANO_ELEICAO')["SIGLA_PARTIDO", "COD_LEGENDA", "NUM_TURNO"].nunique().plot.bar(title=TITULO, grid=False)
plt.show()
```

```python
primeiro_turno[['ANO_ELEICAO', 'COD_LEGENDA', "SIGLA_PARTIDO"]].drop_duplicates().groupby(['ANO_ELEICAO', 'COD_LEGENDA']).count()
```
