# Candidatos das Eleições Municipais - São Paulo

```python
import os
```

```python
import pandas as pd
from IPython.display import IFrame
```

```python
from eleicoes.candidatos import *
```

```python
UF = 'SP'
MUNICIPIO = 71072
ANO_ELEICAO = 2016
```

```python
df = dados_candidatos(ANO_ELEICAO, UF, MUNICIPIO)
```

```python
PREFEITO = 11
VICE_PREFEITO = 12

colunas = ['NUM_TURNO', 'COD_LEGENDA', 'SIGLA_PARTIDO', 'NOME_CANDIDATO', 'DESC_CARGO',
           'COD_SIT_TOT_TURNO','DESC_SIT_TOT_TURNO']

nomes = df.query('COD_CARGO in (@PREFEITO, @VICE_PREFEITO)')[colunas]
nomes = nomes.sort_values(['NUM_TURNO', 'COD_LEGENDA','DESC_CARGO']).reset_index(drop=True)
```

```python
nomes
```

```python
qtd_turnos = len(nomes['NUM_TURNO'].unique())
print(f'Eleição realizada em {qtd_turnos} turnos.')
```

```python
if qtd_turnos == 1:
    turno1 = nomes
else:
    turno1, turno2 = [df for _, df in nomes.groupby('NUM_TURNO')]
```

```python
turno1
```

```python
qtd_legendas = turno1['COD_LEGENDA'].drop_duplicates().count()
print(f'{qtd_legendas} legendas participaram do primeiro turno das Eleições Municipais de {MUNICIPIO}.')

qtd_partidos = turno1['SIGLA_PARTIDO'].drop_duplicates().count()
print(f'{qtd_partidos} partidos participaram do primeiro turno das Eleições Municipais de {MUNICIPIO}.')
```

```python
chapas_unicas = list(filter(lambda x: len(x) == 1, turno1.groupby('COD_LEGENDA')['SIGLA_PARTIDO'].unique()))
txt = ', '.join(i[0] for i in chapas_unicas)
print(f'{len(chapas_unicas)} partidos ({txt}) lançaram chapas únicas.')
```
